```python
import sempy.fabric as fabric 

dataset = "SalesOrders"
workspace = "Fabric PBI Portugal - Semantic Link"
tables = fabric.list_tables(dataset)
tables
```

```python
datasets = fabric.list_datasets()
datasets
```

```python
partitions = fabric.list_partitions(dataset)
partitions = partitions.drop(columns=['Description', 'Error Message', 'Data View', 'Source Type', 'Query Group'])
partitions
```

```python
# Objects to refresh, define using a dictionary
objects_to_refresh = [
    {
        "table": "Sales Invoices",
        "partition": "Sales Invoices"
    }
]

# Refresh the dataset
fabric.refresh_dataset(
    dataset=dataset, 
    objects=objects_to_refresh
)

# List the refresh requests
fabric.list_refresh_requests(dataset=dataset)
```
